#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

# Importing our libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading our datasets

```
url='http://bit.ly/autolib_dataset'
paris = pd.read_csv(url)
paris.tail()
```

# Data Cleaning

```
paris.isnull().sum()
```

```
paris['City'].unique()
```

```
ideal_set =paris[paris.City == 'Paris']
```

```
ideal_set.head()
```

```
ideal_set.head()
```

```
ideal_set.info()
```

Dropping columns that are not needed

```
ideal_set.drop(['Cars', 'month','year','City','Displayed comment',
                'Scheduled at','Geo point','Public name',
                'Station type','Postal code','Address','ID','Charge Slots','Kind'], axis = 1,inplace= True)
```

```
ideal_set.drop(['minute'],axis=1,inplace= True)
```

```
ideal_set.head()
```

```
ideal_set.isnull().any()
```

```
ideal_set.isnull().sum()
```

# Data Analysis

```
ideal_set.groupby('hour')['Bluecar counter'].sum().sort_values(ascending= False)
```

```
available1=ideal_set['Rental status']=='operational'
```

```
available1.unique()
```

```
available2 = ideal_set['Status']== 'ok'
```

```
available2.unique()
```

```
available3 = ideal_set['Subscription status']=='operational'
```

```
available3.unique()
```

```
ideal_set[available2 & available1 & available3]
```

```
 paris_available = ideal_set[available2 & available1 & available3]
```

```
paris_available.head()
```

```
paris_available.groupby('hour')['Bluecar counter'].sum()
```
